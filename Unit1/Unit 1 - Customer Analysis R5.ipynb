```python
# To split columns with numerical values from the categorical ones
# First we get the numeric columns in X_num
# Then we define a new list with the categorical columns: the columns of X
# dropping the columns of X_num

X_num = X._get_numeric_data()

X_cat = X.drop(columns = X_num.columns)
```

```python
# normalize - It is a skewed graph and we want to get a normal distribution

from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X_num)

X_scaled
```

```python


def normalize(X):
    X_mean=X.mean(axis=0)
    X_std=X.std(axis=0)
    X_std[X_std==0]=1.0
    X=(X-X_mean)/X_std
    return X

X_num = normalize(X_num)
X_num.head()
```
